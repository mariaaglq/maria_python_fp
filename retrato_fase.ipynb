```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el sistema: x' = 2*x - 3*y,  y' = -x
def sistema(x, y):
    dx = 2 * x - 3 * y
    dy = - x
    return dx, dy

# Creamos una malla de puntos en el plano
x_min, x_max = -10, 10
y_min, y_max = -10, 10
x_vals = np.linspace(x_min, x_max, 20)
y_vals = np.linspace(y_min, y_max, 20)
X, Y = np.meshgrid(x_vals, y_vals)

# Calculamos el campo vectorial
U, V = sistema(X, Y)

# Configuramos la figura
plt.figure(figsize=(8, 8))

# Dibujamos el campo de direcciones usando streamplot
plt.streamplot(X, Y, U, V, density=1.2, color='blue', arrowstyle='->', arrowsize=1.5)

# Etiquetas y ttulo
plt.xlabel('x')
plt.ylabel('y')
plt.title('Retrato de fase del sistema: x\' = 2x - 3y,  y\' = -x')
plt.grid(True)

# Dibujamos los ejes
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)

# Trazamos las variedades (eigen-direcciones) como lneas punteadas
# Variedad inestable: asociada a lambda=3 y autovector (-3,1) => y = -1/3 x
x_line = np.linspace(x_min, x_max, 100)
plt.plot(x_line, -1/3 * x_line, 'r--', linewidth=2, label='Variedad inestable: y = -1/3 x')

# Variedad estable: asociada a lambda=-1 y autovector (1,1) => y = x
plt.plot(x_line, x_line, 'g--', linewidth=2, label='Variedad estable: y = x')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definicin del sistema de ecuaciones
def sistema(x, y):
    dx = 3*x + y
    dy = 2*x + 2*y
    return dx, dy

# Creacin de una malla para visualizar el campo en el plano (x,y)
x_min, x_max = -10, 10
y_min, y_max = -10, 10
x_vals = np.linspace(x_min, x_max, 20)
y_vals = np.linspace(y_min, y_max, 20)
X, Y = np.meshgrid(x_vals, y_vals)

# Clculo de los valores del campo vectorial en la malla
U, V = sistema(X, Y)

# Configuracin de la figura
plt.figure(figsize=(8, 8))

# Dibujar el campo vectorial con streamplot
plt.streamplot(X, Y, U, V, density=1.2, color='blue', arrowstyle='->', arrowsize=1.5)

# Etiquetas y ttulo
plt.xlabel('x')
plt.ylabel('y')
plt.title("Retrato de fase del sistema:\n x' = 3x + y,   y' = 2x + 2y")
plt.grid(True)

# Dibujar los ejes
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)

# Trazado de las direcciones de los autovectores
x_line = np.linspace(x_min, x_max, 100)

# Para  = 1, autovector (1,-2)  => direccin: y = -2x
plt.plot(x_line, -2*x_line, 'r--', linewidth=2, label='Direccin eigenvector: y = -2x ( = 1)')

# Para  = 4, autovector (1,1)  => direccin: y = x
plt.plot(x_line, x_line, 'g--', linewidth=2, label='Direccin eigenvector: y = x ( = 4)')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definicin del sistema de ecuaciones
def sistema(x, y):
    dx = 2*x - y
    dy = x + 4*y
    return dx, dy

# Creacin de una malla para visualizar el campo en el plano (x,y)
x_min, x_max = -10, 10
y_min, y_max = -10, 10
x_vals = np.linspace(x_min, x_max, 20)
y_vals = np.linspace(y_min, y_max, 20)
X, Y = np.meshgrid(x_vals, y_vals)

# Clculo de los valores del campo vectorial en la malla
U, V = sistema(X, Y)

# Configuracin de la figura
plt.figure(figsize=(8, 8))

# Dibujar el campo vectorial con streamplot
plt.streamplot(X, Y, U, V, density=1.2, color='blue', arrowstyle='->', arrowsize=1.5)

# Etiquetas y ttulo
plt.xlabel('x')
plt.ylabel('y')
plt.title("Retrato de fase del sistema:\n x' = 2x - y,   y' = x + 4y")
plt.grid(True)

# Dibujar los ejes
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)

# Trazado de las direcciones de los autovectores
x_line = np.linspace(x_min, x_max, 100)

# Para  = 1, autovector (1,-2)  => direccin: y = -2x
plt.plot(x_line, -x_line, 'r--', linewidth=2, label='Direccin eigenvector: y = -x ( = 2)')

# Para  = 4, autovector (1,1)  => direccin: y = x
plt.axvline(x=0, color='g', linestyle='--', linewidth=2, label='Direccin eigenvector: x = 0 ( = 0)')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir un rango para x
x = np.linspace(-10, 10, 400)

# Recta en la direccin del vector (2,1): y = x/2
y_line = x / 2

plt.figure(figsize=(8, 6))

# Graficar la recta y = x/2
plt.plot(x, y_line, label='Recta en direccin (2,1): y = x/2', color='blue')

# Graficar la recta en la direccin del vector (0,1), es decir, x=0 (eje y)
plt.axvline(0, color='red', label='Recta en direccin (0,1): x = 0')

# Agregar detalles al grfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Rectas representando los vectores (2,1) y (0,1)')
plt.legend()
plt.grid(True)
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos un rango razonable para x e y
x_vals = np.linspace(1, 3, 400)  # x >= 1
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)

# Definimos la condicin del dominio: y >= 2 - x^2
region = Y >= 2 - X**2

plt.figure(figsize=(8, 6))

# Graficamos la frontera de la parbola: y = 2 - x^2
x_line = np.linspace(1, 3, 400)
y_line = 2 - x_line**2
plt.plot(x_line, y_line, 'r-', linewidth=2, label=r'$y=2-x^2$ (frontera)')

# Graficamos la lnea vertical x = 1
plt.axvline(x=1, color='b', linestyle='--', linewidth=2, label=r'$x=1$')

# Rellenamos la regin que satisface ambas desigualdades
plt.contourf(X, Y, region, levels=[0.5, 1.5], alpha=0.3, colors=['orange'])

# Ajustamos la grfica
plt.xlim(0.5, 3)
plt.ylim(-3, 3)
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Dominio: $x^2+y \geq 2$ y $x \geq 1$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
x0_values = [-2, 0, 4]  # Diferentes condiciones iniciales
t = np.arange(0, 10, 1)  # Tiempo de 0 a 10

# Funcin para calcular x_t
def x_t(t, x0):
    C = x0 - 2
    return C * (-0.5)**t + 2

# Graficar para diferentes valores de x0
plt.figure(figsize=(8, 5))

for x0 in x0_values:
    plt.plot(t, x_t(t, x0), marker='o', label=f'$x_0 = {x0}$')

# Etiquetas y ttulo
plt.axhline(y=2, color='gray', linestyle='--', label='Equilibrio $x=2$')
plt.xlabel('t')
plt.ylabel('$x_t$')
plt.title('Soluciones de la ecuacin en diferencias')
plt.legend()
plt.grid(True)

plt.show()
```
